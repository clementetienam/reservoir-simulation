```python
# Initialization
tstep = 10  # initial time-step in days
tmax = 1500  # end of simulation in days
sim_time = prop_time(tstep=tstep,
                     tmax=tmax)
grid = prop_grid(Nx=23,
                 Ny=23,
                 Nz=1)  # no of grid blocks in x, y, and z direction
res = prop_res(Lx=6900,  # reservoir length in ft
               Ly=6900,  # reservoir width in ft
               Lz=100,  # reservoir height in ft
               press_n=np.ones(grid.Nx * grid.Ny) * 4500,  # block pressure at time level n in psi
               sg_n=np.ones(grid.Nx * grid.Ny) * 0,  # gas saturation at time level n
               press_n1_k=np.ones(grid.Nx * grid.Ny) * 4500,  # block pressure at time level n+1, iteration k
               sg_n1_k=np.ones(grid.Nx * grid.Ny) * 0)  # gas saturation at time level n+1, iteration k
rock = prop_rock(kx=np.ones(grid.Nx * grid.Ny) * 80,  # permeability in x direction in mD
                 ky=np.ones(grid.Nx * grid.Ny) * 120,  # permeability in y direction in mD
                 por=0.22,  # porosity in fraction
                 cr=0)  # 1/psi
fluid = prop_fluid(c_o=0.8e-5,  # oil compressibility in 1/psi
                   mu_o=2.5,  # oil viscosity in cP
                   rho_o=49.1,  # lbm/ft3
                   p_bub=3400,  # pb in psi
                   p_atm=14.7)  # atmospheric pressure in psi
well1 = prop_well(loc=(12, 12),  # well location
                  q=3000,  # well flowrate in STB/D
                  q_lim=3000,  # well flowrate limit in STB/D
                  pwf=2000,  # well BHP in psi
                  pwf_lim=2000,  # well BHP limit in psi
                  rw=0.25,  # well radius in ft
                  qo_control=False)
props = {'rock': rock, 'fluid': fluid, 'grid': grid, 'res': res, 'well': [well1], 'time': sim_time}

# Check Fluid and Rock Properties
fluid.plot_all()
rock.plot_all()

# Distribute properties (and their values) to the grid
params = distribute_properties(props)
mat = np.reshape(np.arange(1, grid.Ny * grid.Nx + 1), (grid.Ny, grid.Nx))
```
